## 13. Business Hours Analysis

```python
if 'hours' in df.columns:
    non_null_hours = df[df['hours'].notnull()]

non_null_hours['hours'].head(3)
```

```python
def has_hours_for_day(hours_dict, day):
    if hours_dict is None or not isinstance(hours_dict, dict):
        return False
    return day in hours_dict

days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

for day in days_of_week:
    non_null_hours[f'open_{day.lower()}'] = non_null_hours['hours'].apply(
        lambda x: has_hours_for_day(x, day) if isinstance(x, dict) else False
    )

for day in days_of_week:
    non_null_hours[f'open_{day.lower()}'] = non_null_hours[f'open_{day.lower()}'].astype(int)

non_null_hours['days_open'] = non_null_hours[[f'open_{day.lower()}' for day in days_of_week]].sum(axis=1)

plt.figure(figsize=(10, 6))
days_counts = non_null_hours['days_open'].value_counts().sort_index()
days_counts.plot(kind='bar')
plt.title('Distribution of Number of Days Open per Week')
plt.xlabel('Number of Days Open')
plt.ylabel('Count of Businesses')
plt.xticks(rotation=0)
plt.show()
```

```python
days_open_data = []
for day in days_of_week:
    count = non_null_hours[f'open_{day.lower()}'].sum()
    percentage = count / len(non_null_hours) * 100
    days_open_data.append({'day': day, 'count': count, 'percentage': percentage})

days_open_df = pd.DataFrame(days_open_data)

plt.figure(figsize=(12,6))
sns.barplot(data=days_open_df, x='day', y='percentage')
plt.title('Percentage of Businesses Open by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Percentage of Businesses Open')
plt.ylim(0, 100)
for i, row in enumerate(days_open_df.itertuples()):
    plt.text(i, row.percentage + 1, f"{row.percentage:.1f}%", ha='center', fontsize=10)
plt.show()
```

```python
def extract_attribute_value(attr_dict, key):
    """Extract attribute value from the attributes dictionary"""
    if attr_dict is None or not isinstance(attr_dict, dict):
        return None
    if key in attr_dict:
        val = attr_dict[key]
        if val == 'True':
            return True
        elif val == 'False':
            return False
        return val
    return None

businesses_with_attrs = df[df['attributes'].notnull()].copy()
businesses_with_attrs['accepts_credit_cards'] = businesses_with_attrs['attributes'].apply(
    lambda x: extract_attribute_value(x, 'BusinessAcceptsCreditCards')
)
businesses_with_attrs['price_range'] = businesses_with_attrs['attributes'].apply(
    lambda x: extract_attribute_value(x, 'RestaurantsPriceRange2')
)
businesses_with_attrs['by_appointment'] = businesses_with_attrs['attributes'].apply(
    lambda x: extract_attribute_value(x, 'ByAppointmentOnly')
)
businesses_with_attrs['accepts_reservations'] = businesses_with_attrs['attributes'].apply(
    lambda x: extract_attribute_value(x, 'RestaurantsReservations')
)

credit_card_stats = businesses_with_attrs.groupby('accepts_credit_cards')['stars'].agg(['mean', 'count'])
print("Average rating by credit card acceptance:")
credit_card_stats
```

## 17. Advanced Attributes Analysis

```python
def parse_nested_attributes(attr_string):
    """Parse attributes that might be nested in string format"""
    if not isinstance(attr_string, str):
        return None
    if attr_string.startswith('{') and attr_string.endswith('}'):
        attr_string = attr_string.replace("'", '"')
        try:
            return json.loads(attr_string)
        except:
            return None
    return attr_string

businesses_with_attrs = df[df['attributes'].notnull()].copy()

businesses_with_attrs['wifi'] = businesses_with_attrs['attributes'].apply(
    lambda x: extract_attribute_value(x, 'WiFi')
)
businesses_with_attrs['wifi'] = businesses_with_attrs['wifi'].apply(
    lambda x: parse_nested_attributes(x) if isinstance(x, str) else x
)

businesses_with_attrs['noise_level'] = businesses_with_attrs['attributes'].apply(
    lambda x: extract_attribute_value(x, 'NoiseLevel')
)

businesses_with_attrs['parking'] = businesses_with_attrs['attributes'].apply(
    lambda x: extract_attribute_value(x, 'BusinessParking')
)

businesses_with_attrs['alcohol'] = businesses_with_attrs['attributes'].apply(
    lambda x: extract_attribute_value(x, 'Alcohol')
)

print("Number of businesses with attribute data:", len(businesses_with_attrs))
businesses_with_attrs[['wifi', 'noise_level', 'alcohol']].head(10)
```

```python
def clean_wifi(wifi_val):
    if wifi_val is None:
        return 'unknown'
    if isinstance(wifi_val, str):
        if 'free' in wifi_val.lower():
            return 'free'
        if 'no' in wifi_val.lower():
            return 'none'
        if 'paid' in wifi_val.lower():
            return 'paid'
    return 'unknown'

businesses_with_attrs['wifi_clean'] = businesses_with_attrs['wifi'].apply(clean_wifi)

wifi_stats = businesses_with_attrs.groupby('wifi_clean').agg({
    'stars': 'mean',
    'business_id': 'count'
}).rename(columns={'business_id': 'count'}).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=wifi_stats, x='wifi_clean', y='stars')
plt.title('Average Rating by WiFi Availability')
plt.xlabel('WiFi Availability')
plt.ylabel('Average Star Rating')
plt.ylim(3.4, 3.8)
for i, row in enumerate(wifi_stats.itertuples()):
    plt.text(i, row.stars + .02, f"{row.count:,}\n({row.stars:.2f})", ha='center')
plt.show()
```

```python
def clean_alcohol(val):
    if val is None:
        return 'unknown'
    if isinstance(val, str):
        if 'full_bar' in val.lower():
            return 'full_bar'
        elif 'beer' in val.lower() or 'wine' in val.lower():
            return 'beer_and_wine'
        elif 'no' in val.lower() or 'none' in val.lower():
            return 'none'
    return 'unknown'

businesses_with_attrs['alcohol_clean'] = businesses_with_attrs['alcohol'].apply(clean_alcohol)

alcohol_stats = businesses_with_attrs.groupby('alcohol_clean').agg({
    'stars': 'mean',
    'business_id': 'count',
    'review_count': 'mean'
}).rename(columns={'business_id': 'count', 'review_count': 'avg_review_count'}).reset_index()

alcohol_stats['percentage'] = alcohol_stats['count'] / alcohol_stats['count'].sum() * 100

order = ['full_bar', 'beer_and_wine', 'none', 'unknown']
alcohol_stats_ordered = alcohol_stats.set_index('alcohol_clean').loc[order].reset_index()

plt.figure(figsize=(10, 6))
plt.bar(alcohol_stats_ordered['alcohol_clean'], alcohol_stats_ordered['stars'], color='skyblue')
plt.title('Average Rating by Alcohol Service Availability')
plt.xlabel('Alcohol Service Availability')
plt.ylabel('Average Star Rating')
plt.ylim(3.0, 4.0)  

for i, row in enumerate(alcohol_stats_ordered.itertuples()):
    plt.text(i, row.stars + 0.02, f"{row.count:,}\n({row.stars:.2f})", ha='center')

plt.tight_layout()
plt.show()
```

## 18. Category Co-occurrence Analysis

```python
from itertools import combinations
from collections import Counter

def get_category_pairs(categories_str):
    if not isinstance(categories_str, str):
        return []
    
    categories = categories_str.split(', ')
    if len(categories) < 2:
        return []
    
    return list(combinations(sorted(categories), 2))

all_pairs = []
for cats in df['categories'].dropna():
    all_pairs.extend(get_category_pairs(cats))

pair_counts = Counter(all_pairs)

top_pairs = pd.DataFrame(pair_counts.most_common(20), columns=['category_pair', 'count'])

print("Most common category combinations:")
top_pairs
```

```python
plt.figure(figsize=(14, 8))
top_pairs['pair_name'] = top_pairs['category_pair'].apply(lambda x: f"{x[0]} & {x[1]}")
sns.barplot(data=top_pairs, x='count', y='pair_name', palette='viridis')
plt.title('Top 20 Most Common Category Combinations', fontsize=16)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Category Pair', fontsize=14)
plt.tight_layout()
plt.show()
```

## 19. Spatial Clustering Analysis

```python
from sklearn.cluster import DBSCAN
import numpy as np

city_name = 'New Orleans'  
city_df = df[df['city'] == city_name].copy()

coords = city_df[['latitude', 'longitude']].values

epsilon_km = 0.3  # 500 meters
epsilon = epsilon_km / 6371.0  # Convert to radians

# Perform DBSCAN clustering
db = DBSCAN(eps=epsilon, min_samples=20, algorithm='auto', metric='haversine').fit(np.radians(coords))

# Add cluster labels to the dataframe
city_df['cluster'] = db.labels_

# Count businesses in each cluster
cluster_counts = city_df['cluster'].value_counts().sort_index()
print(f"Number of clusters found: {len(cluster_counts) - (1 if -1 in cluster_counts.index else 0)}")
print(f"Number of businesses not assigned to any cluster: {cluster_counts.get(-1, 0)}")

# Show the largest clusters
largest_clusters = cluster_counts[cluster_counts.index >= 0].sort_values(ascending=False).head(10)
print("\nTop 10 largest clusters:")
print(largest_clusters)
```

```python
# Visualize businesses on a map, colored by cluster
import folium
from folium.plugins import MarkerCluster

# Get mean coordinates to center the map
center_lat = city_df['latitude'].mean()
center_lon = city_df['longitude'].mean()

# Create a map centered on the city
m = folium.Map(location=[center_lat, center_lon], zoom_start=12)

# Add businesses to the map, colored by cluster
# Only include businesses that are part of a cluster (cluster >= 0)
clustered_businesses = city_df[city_df['cluster'] >= 0]

# Set up a color map for clusters
import matplotlib.cm as cm
import matplotlib.colors as colors

# Get unique cluster values and create a normalized color map
clusters = sorted(clustered_businesses['cluster'].unique())
norm = colors.Normalize(vmin=min(clusters), vmax=max(clusters))
cmap = cm.get_cmap('tab20', len(clusters))

# Create a marker cluster group
marker_cluster = MarkerCluster().add_to(m)

# Add each business to the map
# To avoid overwhelming the notebook, sample 1000 businesses at most
sample_size = min(1000, len(clustered_businesses))
sampled_businesses = clustered_businesses.sample(n=sample_size, random_state=42) if sample_size < len(clustered_businesses) else clustered_businesses

for idx, row in sampled_businesses.iterrows():
    # Get color based on cluster
    cluster_color = colors.rgb2hex(cmap(norm(row['cluster'])))
    
    # Create popup content
    popup_content = f"""
    <b>{row['name']}</b><br>
    Rating: {row['stars']} stars<br>
    Reviews: {row['review_count']}<br>
    Categories: {row['categories']}<br>
    Cluster: {row['cluster']}
    """
    
    # Add marker to the map
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=5,
        popup=folium.Popup(popup_content, max_width=300),
        color=cluster_color,
        fill=True,
        fill_color=cluster_color
    ).add_to(marker_cluster)

# Display the map
m
```

```python
# Find the most common categories in each major cluster
def get_top_categories(cluster_id, top_n=5):
    """Get top categories for a given cluster""",
    # Filter businesses in this cluster
    cluster_businesses = city_df[city_df['cluster'] == cluster_id]
    
    # Explode categories
    all_categories = []
    for cat_str in cluster_businesses['categories'].dropna():
        if isinstance(cat_str, str):
            all_categories.extend(cat_str.split(', '))
    
    # Count category frequencies
    cat_counts = Counter(all_categories)
    return cat_counts.most_common(top_n)

# Get top categories for the largest clusters
top_clusters = cluster_stats.head(5)['cluster'].values

print("Top categories in major business clusters:")
for cluster_id in top_clusters:
    top_cats = get_top_categories(cluster_id)
    print(f"\nCluster {cluster_id} (Size: {cluster_counts[cluster_id]})")
    for cat, count in top_cats:
        print(f"  {cat}: {count} businesses")
```

## 20. Business Success Factors Analysis

### A *Success Score* is created:

success = stars \* (is_open + 0.5) <--- weighs open businesses higher than closed

Businesses with a success score $\geq$ 75 percentile are considered successful

```python
df['success_score'] = df['stars'] * (df['is_open'] + 0.5) 

plt.figure(figsize=(10, 6))
sns.histplot(df['success_score'], bins=7)
plt.title('Distribution of Business Success Score')
plt.xlabel('Success Score')
plt.ylabel('Count')
plt.show()

success_threshold = df['success_score'].quantile(0.75)
df['is_successful'] = df['success_score'] >= success_threshold

print(f"Success threshold: {success_threshold:.2f}")
print(f"Number of successful businesses: {df['is_successful'].sum()} ({df['is_successful'].mean()*100:.1f}%)")
```

```python
categories_exploded = df.dropna(subset=['categories']).copy()
categories_exploded['category'] = categories_exploded['categories'].str.split(', ')
categories_exploded = categories_exploded.explode('category')

category_success = categories_exploded.groupby('category').agg({
    'business_id': 'count',
    'is_successful': 'mean'
}).rename(columns={'business_id': 'count', 'is_successful': 'success_rate'})

significant_cats = category_success[category_success['count'] >= 100].sort_values('success_rate', ascending=False)

print("Categories with highest success rates:")
print(significant_cats.head(15))

print("\nCategories with lowest success rates:")
print(significant_cats.tail(15))
```

```python
plt.figure(figsize=(14, 10))

plt.subplot(2, 1, 1)
top_success_cats = significant_cats.head(10).reset_index()
sns.barplot(data=top_success_cats, y='category', x='success_rate', palette='viridis', hue='category')
plt.title('Categories with Highest Business Success Rate', fontsize=14)
plt.xlabel('Success Rate', fontsize=12)
plt.ylabel('Category', fontsize=12)

plt.subplot(2, 1, 2)
bottom_success_cats = significant_cats.tail(10).reset_index()
sns.barplot(data=bottom_success_cats, y='category', x='success_rate', palette='viridis', hue='category')
plt.title('Categories with Lowest Business Success Rate', fontsize=14)
plt.xlabel('Success Rate', fontsize=12)
plt.ylabel('Category', fontsize=12)

plt.tight_layout()
plt.show()
```
